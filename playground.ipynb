```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('librosa_signal.json') as f:
   data = json.load(f)


with open('indata.json') as f:
   stream_data = json.load(f)
   stream_data = stream_data['indata']
   stream_data = [x[0] for x in stream_data]
```

```python
print(len(np.unique(data)))
print(min(data))
print(max(data))
# len(data)
```

```python
sr = 44000
ft = np.fft.fft(data)
magnitude_spectrum = np.abs(ft)
print(len(magnitude_spectrum))
frequency = np.linspace(0, sr, int(len(magnitude_spectrum)/2))
# num_frequency_bins = int(len(frequency) * f_ratio)
# frequency = [x for x in frequency if x < freq_filter]
magnitude_spectrum = magnitude_spectrum[:len(frequency)]
print(len(magnitude_spectrum))
plt.figure(figsize=(10,6))
plt.plot(frequency, magnitude_spectrum)
# plt.plot(frequency[:num_frequency_bins], magnitude_spectrum[:num_frequency_bins])
plt.xlabel('Frequency [Hz]')
plt.ylabel('Magnitude')
plt.show()
```

```python
hamming = stream_data * np.hamming(len(stream_data))
fft_stream_data = np.abs(np.fft.fft(hamming))
plt.plot(fft_stream_data[:int(len(fft_stream_data)/2)])
# print(hamming)
```
